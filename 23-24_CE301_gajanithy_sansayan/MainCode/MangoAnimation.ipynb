```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from spectral import imshow, kmeans
from spectral.io import envi
import numpy as np
from mango_Data_front import *
from mango_Data_back import *
from IPython.display import HTML
from PIL import *

mango_data = []

# Function which updates the plot frame by frame
def draw_animation(frame): 
    
    plt.clf()
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Reflectance")
    plt.ylim(0,1)

    current_day_index = frame 
    # as day 6 is missing setting the day index to the actual day of the image
    if (current_day_index >= 5):
        day_index = current_day_index +2
    else:
        day_index = current_day_index +1
    # extracting image data for kmeans algorithm
    current_data = mango_data[current_day_index]

    (a,a1) = kmeans(current_data, Nclusters, 60)
    title = f'Clusters - Day {day_index} '
    plt.title(title)

    # Plot the clusters for the current iteration
    for i in range(a1.shape[0]):
        plt.plot(a1[i], label=f'Cluster {i+1}')

    
    plt.legend()
    plt.grid() 

#this is draw animation of spectral data recorded of the back of the fruit.
# Set up the figure and initial plot
graph_front, sp = plt.subplots()

Nclusters = 5

# Creating the animation
total_frames = 11
mango_data = Mango_Front_data
ani = FuncAnimation(graph_front, draw_animation, frames=total_frames, interval=500, repeat=False)
# saving the animation as gif
ani.save('/home/sayan/ce301/coding/mango_data_animation/MangoFront_data.gif', writer='Pillow')
# Converting the animation to HTML for inline display
HTML(ani.to_jshtml())

```

```python
#this is draw animation of spectral data recorded of the back of the fruit.
graph_back, sp = plt.subplots()

Nclusters = 5

# Creating the animation
total_frames = 11
mango_data = Mango_Back_data
ani = FuncAnimation(graph_back, draw_animation, frames=total_frames, interval=500, repeat=False)
# saving the animation as gif
ani.save('MangoBack_data.gif', writer='Pillow')
# Converting the animation to HTML for inline display
HTML(ani.to_jshtml())
```
